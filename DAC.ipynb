# **(二) Numpy 產生隨機亂數**

產生隨機亂數目的：用於統計模擬、最佳化運算解根穩定性、產生假資料、對應真實應用情境產生對應數值...等。

所有隨機亂數基本上都是基於均勻分配 (Uniform distribution) 去產生的。
若要產生其他特定機率分配 (Probability Distribution) 的隨機亂數，也是基於先產生均勻分配後，再去做一些機率反函數的映射 (mapping) 來得到特定機率分配的隨機亂數，此理論即為 **Probabiltiy transformation theorem** (有興趣自己去看)

下圖左邊即為均勻分配，右邊為常態分配。

均勻分配可以想像成 : 就是產生 0-1 之間的隨機亂數，每個數字被取到的機率都一樣。只是差別在於，今天我的隨機亂數的範圍可以自己決定，例 : 0-2 的隨機亂數、10-100 的隨機亂數...

image.png

(Ref : Mohamed A. A. Abdelkareem et. al. (2020), Monte Carlo Sensitivity Analysis of Vehicle Suspension Energy Harvesting in Frequency Domain)

```python
import numpy as np

#set.seed，用於將隨機亂數固定
np.random.seed(5478)

n = 10 #sample size

random_uniform = np.random.uniform(0, 1, size = n) #產生 n 筆來自 min = 0, max = 1 uniform distribution 的數值

random_normal = np.random.normal(0, 1, size = n) #產生 n 筆來自 mean = 0, sd = 1 Normal distribution 的數值


print(random_uniform)
print(random_normal)
print('Mean of Unifrom : ', round(random_uniform.mean(), 2)) #平均數，且四捨五入到小數點後第二位
print('SD of Uniform : ', round(random_uniform.var(), 2)) #標準差，且四捨五入到小數點後第二位
print('Mean of Standard Normal : ', round(random_normal.mean(), 2)) #平均數，且四捨五入到小數點後第二位
print('SD of Standrad Normal : ', round(random_normal.var(), 2)) #標準差，且四捨五入到小數點後第二位


#思考 : 為何抽樣自 Standard Normal 的數據，平均值為 0.5，變異數為 0.59，並非 0、1 ? 
```

# **(三) Panda 簡介與產生 panda dataframe 方法**

Panda 之名源自於 Python Data Analysis Library，主要是基於 Numpy 的工具，目的是用於處理大量數據分析任務。他主要包含了兩種結構：


1.   Series：與 Numpy 中的 vector (一維 array) 相似，也類似於 Python 本身內建的 list，差異在於 Array 與 Series 只能儲存同樣類型資料，而 list 則沒有這限制。要求儲存相同資料類型使得運算效能與記憶體使用都較佳。
2.   Dataframe：類似關聯式資料庫的 Excel，由 row、column 所組成的表格結構，實際上每一個 column 就是一個 Series，因此也可以理解成：數個 Series 可以組合成一個 Dataframe



![Structure of dataframe](https://bookdata.readthedocs.io/en/latest/_images/base_01_pandas_5_0.png)(Ref:https://bookdata.readthedocs.io/en/latest/base/01_pandas.html)



```python
import pandas as pd
import numpy as np

#set.seed，用於將隨機亂數固定
np.random.seed(5478)

#sample size
n = 5

#Create pd Series
random_number = pd.Series(np.random.uniform(0, 1, n), index=['a', 'b', 'c', 'd', 'e']) #index 為 row index，不需設定 column index，因 pd.Series 是一維的


print(random_number)
print('pd.Series 的第 1 個 element', random_number[0])
print('pd.Series 的 a row index', random_number['a'])
print('pd.Series 的第 1~3 個 element :\n', random_number[0:3])
print('pd.Series 直到 row index 為 c :\n', random_number[:'c'])

#沒有 row index 就沒辦法以 index 叫出來

random_number = pd.Series(np.random.uniform(0, 1, n)) #不設定 row index
print(random_number) #可以發現 row.index 被 default 設定成 0, 1, 2, ..., 5
print(random_number['a']) #此時已沒有 row index =a，因此會回傳 error
```

# **(練習一)：建造隨機亂數的統計量並形成一個 Panda data frame**

題目：請用隨機亂數法分別產生 n = 100、 n = 100,000 的標準常態分配，並分別算出其對應的
平均數與變異數並四捨五入到小數點第五位，隨後將結果儲存於一個 2 x 2 的 dataframe 中，column index 分別是 : mean、variance；row index 分別是：n=100, n=100,000

```python
import numpy as np
import pandas as pd

#固定隨機性
np.random.seed(5478)

#產生兩種 normal distribution
random_normal_small = np.random.normal(0, 1, size = 100)
random_normal_large = np.random.normal(0, 1, size = 100000)

#各自計算兩種 normal distribution 的 mean & variance
mean_small_normal = round(random_normal_small.mean(),5)
mean_large_normal = round(random_normal_large.mean(),5)
var_small_normal = round(random_normal_small.var(),5)
var_large_normal = round(random_normal_large.var(),5)

#建立 mean_vector (mean column) & variance vector (variance column)
mean_vector = [mean_small_normal, mean_large_normal]
var_vector = [var_small_normal, var_large_normal]

#方法 1：以 pd.Series 建立 pd.dataframe
mean_series = pd.Series(mean_vector, index = ['n=100', 'n=100,000'])
variance_series = pd.Series(var_vector, index = ['n=100', 'n=100,000'])

#將 pd.series 合成變成 pd.dataframe，並以 keys 命名 column.index
df_normal_distribution_series = pd.concat([mean_series, variance_series], axis=1, keys= ['mean', 'variance'])

################################################################3

#方法 2：以 List 建立 pd.dataframe
normal_distribution_list = {'mean': mean_vector, #mean.index，column 值 = mean_vector
        'variance': var_vector} #var.index，column 值 = var_vector

#將 list 轉換為 pd.dataframe，並多設立 row.index
df_normal_distribution_list = pd.DataFrame(normal_distribution_list, index=['n=100', 'n=100,000'])

print(df_normal_distribution_series)
print(df_normal_distribution_list)
```

# **(七) Panda column 資料型態檢查**

1.   Float64：具小數點的連續數值，例如 162.5
2.   Int64：整數數值，例如 162
1.   Object：字串型態(建議用於**無限制**字串長度的文字，例如姓名)
1.   Category：字串型態(建議用於**有限制**字串長度的文字，例如類別變項、血型...)
2.   Datetime64：時間資料，例如 2023-03-01


在轉學考資料就有這個狀況，很明顯許多變項實際上是類別變項，但 Python 把他當作連續的數字資料了，所以我們就得修改。

參考：為何 Panda dataframe data type 沒有 string 只有 object : https://reurl.cc/n759An

參考：Dataframe Object & Category 兩者差異與記憶體效率比較：https://reurl.cc/9V5jAv

```python
#重複值

#find duplicate rows across all columns
transfer_data_duplicaterows_all = transfer_data[transfer_data.duplicated()]

#find duplicate rows across specific columns
transfer_data_duplicaterows_specific = transfer_data[transfer_data.duplicated(['tuition', 'tutoring'])]

print(transfer_data_duplicaterows_all)
print(transfer_data_duplicaterows_specific)
```

```python
#遺失值

#把資料中有 NA 的全部刪除
transfer_data_drop_all_na = transfer_data.dropna()

#把特定 column 有 NA 的全部刪除
transfer_data_drop_specific_na = transfer_data.dropna(subset=['hours_of_study_everyday'])

print(transfer_data_drop_all_na.info())
print(transfer_data_drop_specific_na.info())

#根據平均值/中位數進行插補
transfer_data['hours_of_study_everyday'] = transfer_data['hours_of_study_everyday'].fillna(transfer_data['hours_of_study_everyday'].mean())
transfer_data['hours_of_study_everyday'] = transfer_data['hours_of_study_everyday'].fillna(transfer_data['hours_of_study_everyday'].median())

#以"有沒有考上"分組後根據平均值/中位數進行插補
transfer_data['hours_of_study_everyday'] = transfer_data.groupby('result')['hours_of_study_everyday'].fillna(transfer_data['hours_of_study_everyday'].mean())
transfer_data['hours_of_study_everyday'] = transfer_data.groupby('result')['hours_of_study_everyday'].fillna(transfer_data['hours_of_study_everyday'].median())

#以"有沒有考上"分組後根據上一個 row / 下一個 row 的值進行插補
transfer_data['hours_of_study_everyday'] = transfer_data.groupby('result')['hours_of_study_everyday'].ffill() #下一個 row (f=firstword)
transfer_data['hours_of_study_everyday'] = transfer_data.groupby('result')['hours_of_study_everyday'].bfill() #上一個 row (b=backword)
```
